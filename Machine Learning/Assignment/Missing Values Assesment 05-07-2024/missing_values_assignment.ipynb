# 1. Loading Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

# 2. Finding Missing Values

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

# 3. Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar chart of survivors vs non-survivors
survivors = df['Survived'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(x=survivors.index, y=survivors.values)
plt.title('Survivors vs Non-survivors')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart of passenges distribution
classes = df['Pclass'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(classes, labels=classes.index, autopct='%1.1f%%', startangle=140)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Line chart of age distribution
plt.figure(figsize=(10, 5))
sns.lineplot(data=df['Age'].dropna())
plt.title('Age Distribution')
plt.xlabel('Index')
plt.ylabel('Age')
plt.show()
```

```python
# Histogram of fare distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Fare'], bins=20, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
